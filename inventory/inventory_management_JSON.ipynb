```python
import json
import time

# Reading the inventory record from the file
with open("record.json", "r") as fd:
    record = json.load(fd)
sales = ''

# Display the menu
print("------------MENU------------\n")
for key in record.keys():
    print(f"{key} : {record[key]['Name']} | {record[key]['Price']} | {record[key]['quantity']}")
print("----------------------------\n")

# Taking user input
ui_name  = str(input("Enter your name         : "))
ui_email = str(input("Enter your email        : "))
ui_phone = str(input("Enter your phone number : "))
ui_pr    = str(input("Enter Product ID        : "))
ui_qn    = int(input("Enter Quantity          : "))

# Checking inventory and processing the order
if record[ui_pr]["quantity"] >= ui_qn:
    record[ui_pr]["quantity"] -= ui_qn

    print("------------BILL------------")
    print(f"Name            : {record[ui_pr]['Name']}")
    print(f"Price           : {record[ui_pr]['Price']}")
    print(f"Quantity        : {ui_qn}")
    print("----------------------------")
    print(f"Billing Amount  : {record[ui_pr]['Price'] * ui_qn}")
    print("----------------------------")
    sales = '1,' + ui_name + ',' + ui_email + ',' + ui_phone + ',' + ui_pr + ',' + record[ui_pr]["Name"] + ',' + str(ui_qn) + ',' + str(record[ui_pr]["Price"]) + ',' + str(ui_qn * record[ui_pr]["Price"]) + ',' + time.ctime() + '\n'

else:
    print("Sorry, we are not having enough quantity in our inventory")
    print(f"We only have {record[ui_pr]['quantity']} quantity")
    print("----------------------------")
    ch = input("Do you want to purchase the available quantity? Press Y/n: ")

    if ch.lower() == "y":
        print("------------BILL------------")
        print(f"Name            : {record[ui_pr]['Name']}")
        print(f"Price           : {record[ui_pr]['Price']}")
        print(f"Quantity        : {record[ui_pr]['quantity']}")
        print("----------------------------")
        print(f"Billing Amount  : {record[ui_pr]['Price'] * record[ui_pr]['quantity']}")
        print("----------------------------")

        sales = '1,' + ui_name + ',' + ui_email + ',' + ui_phone + ',' + ui_pr + ',' + record[ui_pr]["Name"] + ',' + str(record[ui_pr]["quantity"]) + ',' + str(record[ui_pr]["Price"]) + ',' + str(record[ui_pr]["quantity"] * record[ui_pr]["Price"]) + ',' + time.ctime() + '\n'
        record[ui_pr]["quantity"] = 0
    else:
        print("Thanks for visiting!")

# Writing the updated inventory back to the file
with open("record.json", 'w') as fd:
    json.dump(record, fd)

# Writing the sale record to the sales file
with open("sale.csv", "a") as fd:
    fd.write(sales)

print("----------------------------")
print(" Thanks for your order, Inventory Updated! ")
print("----------------------------")
```
